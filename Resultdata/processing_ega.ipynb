```python
import pandas as pd

# Lese die CSV-Datei "Test1 VB10 EGA v03.csv" in ein Pandas-DataFrame ein
ega = pd.read_csv("Test1 VB10 EGA v03.csv", sep=";", encoding="latin1")
```

```python
ega.rename(columns={"ï»¿Section in EGA": "Section in EGA"}, inplace=True)
```

```python
ega.info()
```

```python
# Lösche alle Spalten hinter Spalte 12
ega = ega.drop(ega.columns[13:], axis=1)
ega.describe()
```

```python
ega.info()
```

```python
ega.head(15)
```

```python
# **** ganz wichtig ❗*****
# Ersetze alle NaN-Werte in der Spalte 'Hazard Type' durch den Wert 'Chapter'
ega['Hazard Type'] = ega['Hazard Type'].fillna('Chapter')
```

```python
ega.head(15)
```

```python
# reduzierter Code, der aber alles ausführt ❗ 
def find_type_and_index(start_index, type):
    for i in range(start_index, 0, -1):
        if ega.iloc[i]["Hazard Type"] == type:
            return i
    return None

index_pairs = []
for i, row in ega.iterrows():
    if row["Hazard Type"] != "Measure":
        continue
    index_measure = i
    # index_sec = find_type_and_index(index_measure, "Section in EGA")
    index_chapter = find_type_and_index(index_measure, "Chapter")
    index_function = find_type_and_index(index_measure, "Function")
    index_hazard = find_type_and_index(index_measure, "Hazard")
    index_cause = find_type_and_index(index_measure, "Cause")
    index_pairs.append((index_chapter, index_function, index_hazard, index_cause, index_measure))

print(index_pairs)
```

```python
# hier die finale Lösung, um die Tabelle zu erstellen
result_table = pd.DataFrame(columns=['Chapter Sec','Chapter ID', 'Chapter Type', 'Chapter Requirement',
                                     'Function Sec','Function ID', 'Function Type', 'Function Requirement',
                                     'Hazard Sec','Hazard ID','Hazard Type', 'Hazard Requirement',
                                     'Cause Sec','Cause ID', 'Cause Type', 'Cause Requirement',
                                     'Measure Sec', 'Measure ID', 'Measure Type', 'Measure Requirement','Measure Description',])

elements = ['Chapter', 'Function', 'Hazard', 'Cause', 'Measure']
              # → sind die Elemente, für die die Angaben in der ega ermittelt werden müssen

for index_tuple in index_pairs:
    row_dict = {}
    for i, element in enumerate(elements):
        sec_key = f"{element} Sec"
        id_key = f"{element} ID"
        type_key = f"{element} Type"
        req_key = f"{element} Requirement"
        if element == "Measure":  
          desc_key = f"{element} Description"
        
        # Spaltenüberschriften in ega:
        sec_value = ega['Section in EGA'][index_tuple[i]]
        id_value = int(ega['ID'][index_tuple[i]])
        type_value = ega['Hazard Type'][index_tuple[i]]
        req_value = ega['Requirement'][index_tuple[i]]
        desc_value = ega['Description'][index_tuple[i]]
        
        row_dict[sec_key] = sec_value
        row_dict[id_key] = id_value
        row_dict[type_key] = type_value
        row_dict[req_key] = req_value
        if element == "Measure":  
          row_dict[desc_key] = desc_value
    
    result_table = result_table.append(row_dict, ignore_index=True)

result_table = result_table.astype({'Chapter ID': 'int32', 'Function ID': 'int32', 'Hazard ID': 'int32', 'Cause ID': 'int32', 'Measure ID': 'int32'})
```
