```python
import csv
import pandas as pd

# *** die csv-Datei muss zuvor in in Excel eingelesen und entsprechend die Spalten getrennt 
#     und wieder als CSV-Datei abgespeichert werden ***

filename = 'Req-Grid VB10 EGA v03 - SubIDs Equals 561938 - Kap04 - 2022-10-17.csv'
# Lese die CSV-Datei in ein Pandas-DataFrame ein
ega = pd.read_csv(filename, sep=';')

#ega.head()
```

```python
# Drucke die aktuelle Spaltenreihenfolge
# print(ega.columns)


# Definiere die neue Spaltenreihenfolge
# new_column_order = ['Project', 'Baseline', 'Type', 'Requirement', 'ID', 'Version', 'Hierarchy #', 'Description', 'Status', 'Priority', 'Owner', 'Owner (First Name)', 'Owner (Last Name)', 'Created By', 'Responsible', 'Registered Interest', 'Created On', 'Last Modified On', 'Modified By', 'Source', 'RequireKey', 'Charm', 'Open Points', 'Hazard Remark', 'Harm', 'AFAP justification', 'Entry Type', 'Kano Ranking', 'Test Level', 'Rearrange Operation', 'Rearrange Reason', 'Hazard', 'Initial Risk Assessment', 'Residual Risk Assessment', 'Hazard Type', 'Validation Procedure State', 'New', 'Export', 'Level of Requirements', 'Level of Test', 'Usability', 'Domain', 'Applicable Generic Hazard', 'Hazard Effect To', 'System S1', 'Target Release S1', 'S1 Subsystem', 'System IGS-Platform', 'Target Release IGS-Platform', 'System PreOPlan', 'Target Release PreOPlan', 'System ISS', 'IGS-Platform Subsystem', 'Target Release ISS', 'System S3', 'Target Release S3', 'S3 Subsystem', 'System NavSim', 'Target Release NavSim', 'Traced System', 'P1 before', 'P2 before', 'P1 after', 'Polarion Migration']

new_column_order = ['Project', 'Baseline', 'Type', 'Hierarchy #','ID', 'Entry Type', 'Hazard Type', 'RequireKey', 'Requirement', 'Description', 'Status', 'Priority', 'Owner', 'Owner (First Name)', 'Owner (Last Name)', 'Created By', 'Responsible', 'Registered Interest', 'Created On', 'Last Modified On', 'Modified By', 'Source', 'Charm', 'Open Points', 'S1 Subsystem', 'Hazard Remark', 'Harm', 'AFAP justification', 'Kano Ranking', 'Test Level', 'Rearrange Operation', 'Rearrange Reason', 'Hazard', 'Initial Risk Assessment', 'Residual Risk Assessment', 'Validation Procedure State', 'New', 'Export', 'Level of Requirements', 'Level of Test', 'Usability', 'Domain', 'Applicable Generic Hazard', 'Hazard Effect To', 'System S1', 'Target Release S1', 'System IGS-Platform', 'Target Release IGS-Platform', 'System PreOPlan', 'Target Release PreOPlan', 'System ISS', 'IGS-Platform Subsystem', 'Target Release ISS', 'System S3', 'Target Release S3', 'S3 Subsystem', 'System NavSim', 'Target Release NavSim', 'Traced System', 'P1 before', 'P2 before', 'P1 after', 'Polarion Migration']


# Ändere die Spaltenreihenfolge im DataFrame
ega_neu = ega.reindex(columns=new_column_order)

# Drucke die neue Spaltenreihenfolge
print(ega_neu.columns)
```

```python
ega['Hazard Type'] = ega['Hazard Type'].fillna('Chapter')
ega = ega.astype({'ID': 'int32'})
ega_neu['Hazard Type'] = ega_neu['Hazard Type'].fillna('Chapter')

# Requirement & Headline in 'Entry Type' durch die Kurzform R, H ersetzen 
ega_neu['Entry Type'] = ega_neu['Entry Type'].apply(lambda x: 'R' if x == 'Requirement' else 'H' if x == 'Headline' else x)
ega_neu = ega_neu.astype({'ID': 'int32'})
```

```python
# Erstelle eine neue Spalte 'Hierarchy', die die 'Hierarchy #' Spalte in eine Liste von Strings aufteilt
ega_neu['Hierarchy'] = ega_neu['Hierarchy #'].str.split('.')

# ega_neu.info()
```

```python
# Drucke die Inhalte der 'Hierarchy' Spalte für die 5. Zeile
# print(ega_neu['Hierarchy'].iloc[4])

# Teile die 'Hierarchy' Spalte für die 5. Zeile in einzelne Elemente auf
hierarchy = ega_neu['Hierarchy'].iloc[4]

# Drucke die Elemente der aufgeteilten 'Hierarchy' Spalte
# print(f"x[0] = {hierarchy[0]}")
# print(f"x[1] = {hierarchy[1]}")
# print(f"x[2] = {hierarchy[2]}")
# print(f"x[3] = {hierarchy[3]}")
# print(f"x[4] = {hierarchy[4]}")
# print(f"x[5] = {hierarchy[5]}")
```

```python
ega_neu['Hierarchy'] = ega_neu['Hierarchy #'].str.split('.').apply(lambda x: x[1:] if len(x) > 0 else None)

ega_neu['Section'] = ega_neu['Hierarchy'].apply(lambda x: x[0] if len(x) > 0 else None)
ega_neu['Chapter'] = ega_neu['Hierarchy'].apply(lambda x: x[1] if len(x) > 1 else None)
ega_neu['Function'] = ega_neu['Hierarchy'].apply(lambda x: x[2] if len(x) > 2 else None)
ega_neu['Hazard'] = ega_neu['Hierarchy'].apply(lambda x: x[3] if len(x) > 3 else None)
ega_neu['Cause'] = ega_neu['Hierarchy'].apply(lambda x: x[4] if len(x) > 4 else None)
ega_neu['Measure'] = ega_neu['Hierarchy'].apply(lambda x: x[5] if len(x) > 5 else None)
```

```python
# ega_neu.info()
ega_neu.drop('Hierarchy', axis=1, inplace=True)
# ega_neu.info()
```

```python
# Bestimme den Index der 'Type' Spalte
type_index = ega_neu.columns.get_loc('Hierarchy #')

# Füge die neuen Spalten zwischen 'Type' und 'ID' ein
ega_neu.insert(type_index + 1, 'Section', ega_neu.pop('Section'))
ega_neu.insert(type_index + 2, 'Chapter', ega_neu.pop('Chapter'))
ega_neu.insert(type_index + 3, 'Function', ega_neu.pop('Function'))
ega_neu.insert(type_index + 4, 'Hazard', ega_neu.pop('Hazard'))
ega_neu.insert(type_index + 5, 'Cause', ega_neu.pop('Cause'))
ega_neu.insert(type_index + 6, 'Measure', ega_neu.pop('Measure'))
```

```python
# Ändere den Typ der Spalten in object
ega_neu['Section'] = ega_neu['Section'].astype('object')
ega_neu['Chapter'] = ega_neu['Chapter'].astype('object')
ega_neu['Function'] = ega_neu['Function'].astype('object')
ega_neu['Hazard'] = ega_neu['Hazard'].astype('object')
ega_neu['Cause'] = ega_neu['Cause'].astype('object')
ega_neu['Measure'] = ega_neu['Measure'].astype('object')

# Wandle alle nicht-None Werte in int um
for column in ['Section', 'Chapter', 'Function', 'Hazard', 'Cause', 'Measure']:
    ega_neu[column] = ega_neu[column].apply(lambda x: int(x) if x is not None else x)

# ega_neu.info()
```

```python
# *** ABSPEICHERN ***
# Verzeichnis bestimmen
import os

# die korrekte Zeitzone definieren, damit die Zeit auch wirklich stimmt
import pytz
import datetime

berlin = pytz.timezone("Europe/Berlin")
now = datetime.datetime.now(tz=berlin)
date_string = now.strftime("%Y-%m-%d - %H-%M")

# Zum abspeichern der Dateien wird aus filename:
filenames = "Req-Grid VB10 EGA v03 --- "
csvfilename = filenames + date_string + ".csv"
xslfilename = filenames + date_string + ".xlsx"

# Pfad zum Unterverzeichnis
subdir = "Resultdata"

# Überprüfen, ob das Unterverzeichnis bereits vorhanden ist
if not os.path.exists(subdir):
    # Erstellen des Unterverzeichnisses, falls es noch nicht vorhanden ist
    os.makedirs(subdir)

# Vollständiger Pfad der Zieldatei
csvfile = os.path.join(subdir, csvfilename)
xlsfile = os.path.join(subdir, xslfilename)

# Schreiben von ega_neu in die CSV-Datei im Unterverzeichnis
ega_neu.to_csv(csvfile, sep=";", encoding="utf-8")

# entfernen der index spalte
ega_neu = ega_neu.reset_index(drop=True)
# ega_neu.to_excel(xlsfile, sheet_name='VB10 EGA v03', index=False)
# ega_neu.to_excel(xlsfile, index=False)

ega_neu = ega_neu.reset_index(drop=True)
ega_neu.to_excel(xlsfile, sheet_name='VB10 EGA v03', index=False)

# ega_neu.head()
```

```python
import openpyxl

# Excel-Datei öffnen
wb = openpyxl.load_workbook(xlsfile)

# Arbeitsblatt auswählen
ws = wb['VB10 EGA v03']
# ws = wb[wb.sheetnames[0]]

# Schriftgröße für den Rest der Tabelle festlegen
for row in ws.rows: 
    for cell in row: 
        cell.font = openpyxl.styles.Font(size=10)

# Schriftgröße für die obere Zeile festlegen
for row in ws['A1:BQ1']: 
    for cell in row: 
        cell.font = openpyxl.styles.Font(size=12)

# Zellenbereich für die obere Zeile auswählen
for row in ws['A1:BQ1']: 
    for cell in row: 
        cell.fill = openpyxl.styles.PatternFill(fgColor='19006B', fill_type='solid')
        cell.font = openpyxl.styles.Font(color='FFFFFF')
```

```python
# Liste der Spalten, die gruppiert werden sollen
columns_to_group = ['Section', 'Chapter', 'Function', 'Hazard', 'Cause', 'Measure', 'Owner (First Name)', 'Owner (Last Name)', 'Created By', 'Responsible', 'Registered Interest', 'Created On', 'Source', 'Charm', 'Open Points', 'Kano Ranking']
```

```python
# Spalten gruppieren
column_indices = {column: index for index, column in enumerate(ega_neu.columns)}
current_group = []
current_group_start = None

for col_name in columns_to_group:
    col_idx = column_indices[col_name]
    if not current_group:
        current_group_start = col_idx + 1 # +1 eingefügt
    current_group.append(col_name)

for col_idx, col_name in enumerate(ega_neu.columns):
    if col_name in columns_to_group:
        if not current_group:
            current_group_start = col_idx + 1  # +1 eingefügt
        current_group.append(col_name)
        
        if col_name != columns_to_group[-1]:
            if ega_neu.columns[col_idx+1] not in columns_to_group:
                ws.column_dimensions.group(start=openpyxl.utils.get_column_letter(current_group_start), end=openpyxl.utils.get_column_letter(current_group_start + len(current_group) - 1), hidden=False)
                # ws.column_dimensions.group(start=openpyxl.utils.get_column_letter(current_group_start+1), end=openpyxl.utils.get_column_letter(current_group_start + len(current_group)), hidden=False)
                current_group = []
    else:
        if current_group:
            ws.column_dimensions.group(start=openpyxl.utils.get_column_letter(current_group_start), end=openpyxl.utils.get_column_letter(current_group_start + len(current_group) - 1), hidden=False)
            # ws.column_dimensions.group(start=openpyxl.utils.get_column_letter(current_group_start), end=openpyxl.utils.get_column_letter(current_group_start + len(current_group)), hidden=False)
            current_group = []

if current_group:
    ws.column_dimensions.group(start=openpyxl.utils.get_column_letter(current_group_start), end=openpyxl.utils.get_column_letter(current_group_start + len(current_group) - 1), hidden=False)
    # ws.column_dimensions.group(start=openpyxl.utils.get_column_letter(current_group_start), end=openpyxl.utils.get_column_letter(current_group_start + len(current_group)), hidden=False)

```
